# **Data Visualization**
Data Visualization is a method of presenting data in graphical or visual form, which facilitates the comprehension and interrpretation of complex information. Given the rising volume and complexity of data generated on a daily basis, the signifigance of effective data visualization has never been greater.

```python
#Import Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Import Dataset into Python using the Pandas Library
daily_customers_df = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletDailyCustomers.csv', index_col=0)
marketing = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletMarketing.csv', index_col=0)
overheads = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletOverheads.csv', index_col=0)
size = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletSize.csv', index_col=0)
staff = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletStaff.csv', index_col=0)
pd.plotting.register_matplotlib_converters()
daily_customers_df.index = pd.to_datetime(daily_customers_df.index)
```

```python
#Merge and Read the Dataset
summary_data = pd.DataFrame(index = daily_customers_df.columns)
summary_data['Daily Customers'] = daily_customers_df.sum().values
summary_data['Marketing'] = marketing.values
summary_data['Overheads'] = overheads.values
summary_data['Outlet Size'] = size.values
summary_data['Staff'] = staff.values
```

# **Exploratory Data Analysis (EDA)**

EDA is a crucial step in the analysis of any dataset. It involves summarizing the main characteristics of the data and identifying patterns, trends, relationships between variables. The purpose of EDA is to gain better understanding of the data and to identify any issues that needs to be addressed before proceeding with further analysis.

```python
daily_customers_df= daily_customers_df.reindex(daily_customers_df.mean().sort_values(ascending= False).index,axis=1)
daily_customers_df
```

```python
plt.figure(figsize=(23,10))
x_pos = np.arange(len(daily_customers_df.columns))
plt.bar(x_pos,daily_customers_df.sum(), align='center')
plt.xticks(x_pos, daily_customers_df.columns)
plt.xlabel('Outlets', fontsize =18)
plt.ylabel('Total number of Daily Customers', fontsize = 18)
plt.title('Total Number of Daily Customers',fontsize =20)
plt.show()
```

```python
categories = ['High','Medium','Low']
categories_selected = [[] for i in range(len(categories))]
for name in daily_customers_df.columns:
  total_hits = daily_customers_df[name].sum()
  if total_hits > 400000:
    category = 0
  elif total_hits > 100000:
    category = 1
  else:
    category = 2
  categories_selected[category].append(name)
  print('Store '+name+' has '+categories[category]+' volume customers')

High_Traffic = categories_selected[0]
Medium_Traffic = categories_selected[1]
Low_Traffic = categories_selected[2]
```

```python
# Define volume categories based on a certain threshold
high_volume = daily_customers_df.columns[daily_customers_df.sum() > daily_customers_df.sum().quantile(0.75)]
medium_volume = daily_customers_df.columns[(daily_customers_df.sum() <= daily_customers_df.sum().quantile(0.75)) & (daily_customers_df.sum() > daily_customers_df.sum().quantile(0.5))]
low_volume = daily_customers_df.columns[daily_customers_df.sum() <= daily_customers_df.sum().quantile(0.5)]

# Filter the data based on the volume category and create a summary table for each category
high_volume_summary = summary_data.loc[high_volume].describe()
medium_volume_summary = summary_data.loc[medium_volume].describe()
low_volume_summary = summary_data.loc[low_volume].describe()

# Identify any anomalies using boxplots
plt.figure()
sns.boxplot(x=summary_data['Daily Customers'])
plt.title('Boxplot of Daily Customers')
plt.show()
```

```python
current_year_outlets = daily_customers_df.columns
previous_year_outlets = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletDailyCustomers.csv', index_col=0).columns

new_outlets = list(set(current_year_outlets) - set(previous_year_outlets))
closed_outlets = list(set(previous_year_outlets) - set(current_year_outlets))

print('New outlets:', new_outlets)
print('Closed outlets:', closed_outlets)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource, HoverTool

output_file("daily_customers.html")

source = ColumnDataSource(daily_customers_df)

p = figure(x_axis_type="datetime", title="Daily Customers for All Outlets")
p.line(x='Date', y='Outlet 1', source=source, color='red', legend_label='Outlet 1')
p.line(x='Date', y='Outlet 2', source=source, color='green', legend_label='Outlet 2')
p.line(x='Date', y='Outlet 3', source=source, color='blue', legend_label='Outlet 3')
p.add_tools(HoverTool(tooltips=[('Date', '@Date{%F}'), ('Customers', '@{Outlet 1}'), ('Customers', '@{Outlet 2}'), ('Customers', '@{Outlet 3}')], formatters={'@Date': 'datetime'}, mode='vline'))

show(p)
```
